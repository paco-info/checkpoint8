```python
# Import des librairies nÃ©cessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, KMeans
from scipy.cluster.hierarchy import dendrogram

# Chargement des donnÃ©es
data = pd.read_csv('CC GENERAL.csv')
data


```

```python
# Suppression de la variable CUST_ID (non nÃ©cessaire pour la segmentation)
data.drop(['CUST_ID'], axis=1, inplace=True)

# Traitement des valeurs manquantes
data.fillna(method='ffill', inplace=True)

# Clustering hiÃ©rarchique
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
hc.fit(data)
```

```python

# Visualisation des clusters
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=hc.labels_, cmap='rainbow')
plt.title('Clustering hiÃ©rarchique')
plt.xlabel('Balance')
plt.ylabel('Purchases')
plt.show()
```

```python

# K-means clustering
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)
```

```python

# Recherche du nombre optimal de clusters
plt.plot(range(1, 11), sse)
plt.title('MÃ©thode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('SSE')
plt.show()
```

```python

# Clustering K-means avec k=4 (nombre optimal de clusters)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(data)
```

```python

# Visualisation des clusters
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=pred_y, cmap='rainbow')
plt.title('Clustering K-means')
plt.xlabel('Balance')
plt.ylabel('Purchases')
plt.show()
```
